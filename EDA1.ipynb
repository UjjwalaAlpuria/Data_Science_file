```python
 #Task 2:
        #STATISTICAL SUMMARY


import pandas as pd

numerical_col=['LB','AC','FM','UC','DL','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP','DS']

summary = pd.DataFrame() # creating a empty data frame for the summery
for col in numerical_col: # using for loop for calculating statistic trough each column values
  summary.loc[col,'Mean'] = df[col].mean() # calculating mean by using location fuction of columns
  summary.loc[col,'Median'] = df[col].median() # calculating median by using location fuction of columns
  summary.loc[col,'Mode'] = df[col].mode()[0] # calculating mode by using location fuction of columns
  summary.loc[col,'Std'] = df[col].std() # calculating standard daviation by using location fuction of columns

  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3-q1

  summary.loc[col, 'iqr'] = iqr

summary =summary.round(2) # displaying the summary in table form
display(summary)
```

```python
# TO FIND OUT IS THERE IS ANY  INTERSTING FINDING  FROM THE SUMMERY

#1.LB has some variation, so it could be useful.

#2.ASTV, MSTV, MLTV, ALTV, and Width show big differences between samples — they might help predict NSP.

#3.AC, FM, UC, DL, DP, DS don’t change at all (all zeros), so they should be removed.

#4.NSP has different classes in the data, which is good for building a model.
```
