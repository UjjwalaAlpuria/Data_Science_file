### Data Preprocessing

Before building the recommendation system, we need to preprocess the data. This involves handling missing values and preparing categorical features like 'genre' for numerical analysis.

```python
# Display basic information and check for missing values
print(anime_df.info())
print("\nMissing values before preprocessing:")
print(anime_df.isnull().sum())
```

#### Handle Missing Values

We will fill missing 'rating' values with the mean, convert 'episodes' to numeric (treating 'Unknown' as NaN and then filling with 1, or median), and fill missing 'genre' values with 'No Genre'.

```python

# Fill missing 'rating' values with the mean
anime_df['rating'].fillna(anime_df['rating'].mean(), inplace=True)

# Convert 'episodes' to numeric, coercing 'Unknown' to NaN, then fill NaN with 1 (common for movies/OVAs with single episode) or median/mean.
anime_df['episodes'] = pd.to_numeric(anime_df['episodes'], errors='coerce')
anime_df['episodes'].fillna(1, inplace=True) # Assuming 'Unknown' often means 1 episode for movies/OVAs, or use median/mean if TV series are more common

# Fill missing 'genre' values with 'No Genre'
anime_df['genre'].fillna('No Genre', inplace=True)

print("\nMissing values after preprocessing:")
print(anime_df.isnull().sum())
```

##Feature Extraction:

Decide on the features that will be used for computing similarity (e.g., genres, user ratings).
Convert categorical features into numerical representations if necessary.
Normalize numerical features if required.




```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity # importing the libiries to perform the task

# 1. Further clean the 'genre' column
# Replace 'No Genre' with an empty string
anime_df['genre'] = anime_df['genre'].replace('No Genre', '')

# Create a CountVectorizer to convert genres into a matrix of token counts
# This effectively performs one-hot encoding for genres that are comma-separated
count_vectorizer = CountVectorizer(tokenizer=lambda x: x.split(', '), analyzer='word')

# Fit and transform the genre column to create the genre matrix
genre_matrix = count_vectorizer.fit_transform(anime_df['genre'])

# Convert the sparse matrix to a DataFrame for better readability and indexing
genre_matrix = pd.DataFrame(genre_matrix.toarray(), columns=count_vectorizer.get_feature_names_out(), index=anime_df['name'])

print("Shape of genre_matrix:", genre_matrix.shape)
print("First 5 rows of genre_matrix:")
print(genre_matrix.head())
```

### Feature Engineering and Normalization

We will now convert the 'type' categorical feature into numerical representation using one-hot encoding. Additionally, we will normalize the numerical features 'rating' and 'members' to ensure they contribute equally to similarity calculations, as their scales can vary significantly.

```python
from sklearn.preprocessing import MinMaxScaler

# One-hot encode 'type' feature
type_dummies = pd.get_dummies(anime_df['type'], prefix='type')
anime_df = pd.concat([anime_df, type_dummies], axis=1)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Normalize 'rating' and 'members'
anime_df['rating_normalized'] = scaler.fit_transform(anime_df[['rating']])
anime_df['members_normalized'] = scaler.fit_transform(anime_df[['members']])

print("DataFrame with new one-hot encoded 'type' columns and normalized 'rating' and 'members':")
display(anime_df.head())
```

Recommendation System:

Design a function to recommend anime based on cosine similarity.
Given a target anime, recommend a list of similar anime based on cosine similarity scores.
Experiment with different threshold values for similarity scores to adjust the recommendation list size.


```python
from sklearn.metrics.pairwise import cosine_similarity

# Ensure genre_matrix index matches anime_df for concatenation
genre_matrix.index = anime_df.index

# Combine all numerical features for similarity calculation
# Selecting relevant columns from anime_df and concatenating with genre_matrix and type_dummies
combined_features = pd.concat([
    genre_matrix,
    type_dummies,
    anime_df[['rating_normalized', 'members_normalized']]
], axis=1)

# Calculate cosine similarity on the combined feature matrix
cosine_sim_combined = cosine_similarity(combined_features)

# Convert the similarity matrix to a DataFrame for easier handling and inspection
cosine_sim_df_combined = pd.DataFrame(cosine_sim_combined, index=anime_df['name'], columns=anime_df['name'])

print("Combined Feature Matrix Shape:", combined_features.shape)
print("Cosine Similarity Matrix (first 5x5) based on combined features:")
display(cosine_sim_df_combined.iloc[:5, :5])
```

##Evaluation:

Split the dataset into training and testing sets.
Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score.
Analyze the performance of the recommendation system and identify areas of improvement.



```python
def recommend_anime(anime_title, cosine_sim_df, anime_df, num_recommendations=10):

    try:
        # Get the similarity scores for the given anime
        sim_scores = cosine_sim_df[anime_title]
    except KeyError:
        print(f"Anime '{anime_title}' not found in the similarity matrix.")
        return []

    # Sort the anime based on the similarity scores
    # Exclude the anime itself (score of 1)
    sim_scores = sim_scores.sort_values(ascending=False)

    # Get the top num_recommendations, excluding the input anime
    top_recommendations = sim_scores.drop(anime_title).head(num_recommendations).index.tolist()

    return top_recommendations

print("The `recommend_anime` function has been defined.")
```

```python
def evaluate_recommendations(anime_title, cosine_sim_df, anime_df, num_recommendations=10, relevance_threshold=0.7):

    try:
        # Get similarity scores for the target anime
        sim_scores = cosine_sim_df[anime_title]
    except KeyError:
        # print(f"Anime '{anime_title}' not found in the similarity matrix. Skipping evaluation.")
        return None, None, None # Return None for metrics if anime not found

    # Identify 'true relevant items' based on the threshold
    # Exclude the anime itself from the true relevant items
    true_relevant_items_series = sim_scores[sim_scores >= relevance_threshold].drop(anime_title, errors='ignore')
    true_relevant_items = set(true_relevant_items_series.index)

    # Get recommendations from the system
    recommended_anime = recommend_anime(anime_title, cosine_sim_df, anime_df, num_recommendations)
    recommended_anime_set = set(recommended_anime)

    # Calculate True Positives (TP)
    # These are items recommended by the system that are also truly relevant
    tp = len(recommended_anime_set.intersection(true_relevant_items))

    # Calculate False Positives (FP)
    # These are items recommended by the system that are not truly relevant
    fp = len(recommended_anime_set.difference(true_relevant_items))

    # Calculate False Negatives (FN)
    # These are truly relevant items that were NOT recommended by the system
    fn = len(true_relevant_items.difference(recommended_anime_set))

    # Calculate Precision, Recall, F1-score
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1_score

print("The `evaluate_recommendations` function has been defined.")

# Select a small, representative sample of anime titles for evaluation
sample_anime_titles = [
    'Naruto',
    'Kimi no Na wa.',
    'Fullmetal Alchemist: Brotherhood',
    'Death Note',
    'Spirited Away',
    'Cowboy Bebop',
    'Mushishi Zoku Shou 2nd Season',
    'Code Geass: Hangyaku no Lelouch R2'
]

# Store evaluation metrics
evaluation_results = []

print("\n--- Starting Evaluation ---")
for title in sample_anime_titles:
    prec, rec, f1 = evaluate_recommendations(title, cosine_sim_df_combined, anime_df)
    if prec is not None:
        evaluation_results.append({'anime': title, 'precision': prec, 'recall': rec, 'f1_score': f1})
        print(f"Evaluating '{title}':")
        print(f"  Precision: {prec:.4f}, Recall: {rec:.4f}, F1-score: {f1:.4f}")
    else:
        print(f"Could not evaluate '{title}'. Anime not found or other issue.")

print("\n--- Evaluation Summary ---")
if evaluation_results:
    avg_precision = np.mean([res['precision'] for res in evaluation_results])
    avg_recall = np.mean([res['recall'] for res in evaluation_results])
    avg_f1_score = np.mean([res['f1_score'] for res in evaluation_results])

    print(f"Average Precision: {avg_precision:.4f}")
    print(f"Average Recall: {avg_recall:.4f}")
    print(f"Average F1-score: {avg_f1_score:.4f}")
else:
    print("No evaluation results to display.")
```
