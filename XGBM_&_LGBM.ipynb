```python
import pandas as pd

# Load the datasets
df_train = pd.read_csv('/content/Titanic_train.csv')
df_test = pd.read_csv('/content/Titanic_test.csv')

# Impute missing values
df_train['Age'].fillna(df_train['Age'].median(), inplace=True)
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)
df_test['Age'].fillna(df_test['Age'].median(), inplace=True)
df_test['Fare'].fillna(df_test['Fare'].median(), inplace=True)

# Dropping columns which are not useful for the model
df_train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df_test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
# Prepare data for Decision Tree
# Assuming 'Survived' is the target variable and the rest are features
X_train = df_train.drop('Survived', axis=1)
y_train = df_train['Survived']
X_test = df_test.copy() # Assuming df_test has the same features as X_train

# Handle categorical features by one-hot encoding
X_train = pd.get_dummies(X_train, columns=['Sex', 'Embarked'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['Sex', 'Embarked'], drop_first=True)

# Align columns - crucial for consistent feature sets
train_cols = X_train.columns
test_cols = X_test.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    X_test[c] = 0

missing_in_train = set(test_cols) - set(train_cols)
for c in missing_in_train:
    X_train[c] = 0

X_test = X_test[train_cols] # Ensure tdt.fithe order of columns is the same
```

```python
 dt = DecisionTreeClassifier(random_state=42)
 dt.fit(X_train, y_train)
```

```python
dt.fit(X_train,y_train)
ypredi=dt.predict(X_test)
# Since there is no y_test for the test set, I will print the classification report and accuracy for the training set.
print(classification_report(y_train,dt.predict(X_train)))
print('Train Accuracy:', dt.score(X_train,y_train))
print('Test Accuracy:',dt.score(X_test,ypredi))
```

```python
lg=LogisticRegression() #using logistic regressing for creating the model
lg.fit(X_train,y_train)
ypred=lg.predict(X_test)
print('Train Accuracy:',lg.score(X_train,y_train))
print('Test Accuracy:',lg.score(X_test,ypred))
```
