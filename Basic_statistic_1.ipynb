```python
# BASIC STATISTICS 1
import pandas as pd  #importing pandas to work on data sets.

df = pd.read_csv('sales_data_with_discounts.csv') #uploading excel files which contain the data set .

print(df.head()) # to view the data

```

```python
df.info() # to identify any null values
```

```python
df.describe() #to identify numerical columns in the data set and it also shows us mean and dtandard daviation of colums.
```

```python
#calculate the meaan , meadia ,mode and standard daviation

#mean () it calculate the average value of all the values of columes
mean_value = df['Net Sales Value'].mean() # we use Net Sales Value because it is more accurate and directly related  to profatibility.
print("Mean value:", mean_value)  # Net Sales Value = Total sales - Discounts- Returns

#median () give the middle value of sorted data
median_value = df['Net Sales Value'].median()
print("Median value:", median_value)

# mode() gve the most frequently occuring values
mode_value = df['Net Sales Value'].mode()[0] # [0] is using for picking up the first values
print("Mode value:", mode_value)

#standard deviation () most variation in sales values
std_value = df['Net Sales Value'].std()
print("Standard Deviation value:", std_value)
```

```python
# DATA VISUALIZATION
# OBJECTIVE= to visualize  the distribution and relationship of numerical amd categorical vaariables in data set

# HISTOGRAM

import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = df.select_dtypes(include='number').columns # for selecting only numerical columns from the data frame
for col in numerical_columns: #for loop use to go through all the numerical columns onr by one
    plt.figure(figsize=(6,4)) #for figure size
    sns.histplot(df[col], bins=20,kde=True) # from seaborn for  drawing a histogram.for specific numerical coloums
                                            # bins=20 for splits the data into 20 intervals
    plt.title(f'Histogram of {col}')# for showing the column name in title
    plt.xlabel(col) #for  showing the column name
    plt.ylabel('Frequency')#for showing  the friquently each range of values appears (frequency)
    plt.show()# for displaying the plot
```

```python
# BOXLPOTS
 #boxplot for numrerical variable to identify outliers and the interquartile range


numerical_columns = df.select_dtypes(include='number').columns  #for selecting only numerical columns
for col in numerical_columns:    #for loop for gooing throught all numerical values one by one to plot a boxplot
    plt.figure(figsize=(6,4))    # for giving plot size
    sns.boxplot(x=df[col],color='lightpink')   # seabourn library for creating a box plot ,df[col] forselecting onr column
                                             #x= for ploting data into x-axis .colour = for seting the colour of box plot
    plt.title(f'Boxplot of {col}')     #set title for the boxplote by using f-string which replace {col} with the name of current column
    plt.xlabel(col)     #set the name for the x-axis for  using the column name
    plt.show()          # for displaying the plot
```

```python
# BAR CHART ANALYSIS FOR CATEGORICAL COLUMN:

# For identify categorical column
# creating bar chart to visualuze the frequency or count of each category
#analyze the distribution of categories and provide insights

categorical_col='Model'   #stores  the name of the categorical column (its a variablr value)

plt.figure(figsize=(10,8))   # providing widht and hight values for graph
sns.countplot(data=df, x=categorical_col, color="lightgreen")  # telling to use seaborn for the data set  and x give the x-axis value wich column to count values from(eg."Model")

plt.title(f'Count of each category in{categorical_col}')   #adding title to the chart
plt.xlabel(categorical_col)   # labling the x-axis using the column name
plt.ylabel('Count')      #labing the y-axis
plt.xticks(rotation=90)   #for rotaing the x-axis lable so long name dont overlap
plt.show()   #for displaying the chart
```

```python
# STANDARDIZATION OF NUMERICAL VARIABLES

#Objective : to scale numerical variable for uniformity ,improving the dataset's suitable for analytical model

import pandas as pd
from sklearn.preprocessing import StandardScaler # import standardization from scikit-learn for standardization.

columns_to_standardize = ['Avg Price', 'Total Sales Value'] #specifying the columns to standardize
df_standardized = df.copy() #making copy fo data frame to avoide cange in original data
scaler = StandardScaler()#creating a standardscale object
df_standardized[columns_to_standardize] = scaler.fit_transform(df_standardized[columns_to_standardize]) #appling scaling to the selected columns and overwriting them in df_standardized

print(df[columns_to_standardize].head()) #displaying the original data
print(df_standardized[columns_to_standardize].head())#data after standardization as resulted output
```

```python
import pandas as pd
import numpy as np

# Assuming 'df' as DataFrame and 'Avg Price' is the column to standardize.

column_name = 'Avg Price'
mu = df[column_name].mean() #calculating mean of the column
sigma = df[column_name].std() #calculating standard daviation of the column

df['Avg Price_Standardized'] = (df[column_name] - mu) / sigma #appling standardization formula for every x value in columns
print("Original 'Avg Price' Column:") # displaying original columns values
print(df[[column_name]])
print("\n" + "="*40 + "\n")# printinf a line of equal signs to visually separate the two output

print("Standardized 'Avg Price' Column:")
print(df[['Avg Price_Standardized']]) #displaying thr result
```

```python
# CONVERSION OF CATEGORICAL DATA INTO DUMMY VARIABLE

#Objective : to transform categorical variable into a format that can be provided to ML algorithims

# we have to first identify categorical columns you want to canvert

import pandas as pd

#list of categorical columns i want to convert
categorical_col=['SKU','City','Volume','Day']
# appling  one hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_col, drop_first=True) #using pd.get.dummies()in pandas to create new columns
print(df_encoded.head())  # Display the transformed dataset
```
