```python
# Capping outliers using the IQR method
for column in df.select_dtypes(include=np.number).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
print("Outliers capped successfully using IQR method.")
```

Feature Scaling
Before applying K-Means, it is good practice to scale the features so that all features contribute equally to the distance calculations. I will use StandardScaler for this purpose.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df.columns = df.columns.astype(str) # Convert all column names to strings
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)

print("Scaled DataFrame head:")
display(df_scaled.head())
```

### K-Means Clustering - Elbow Method


```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

## Implement Hierarchical Clustering

### Subtask:
Perform hierarchical clustering on the scaled dataset using `AgglomerativeClustering` with different linkage criteria.


**Reasoning**:
To implement hierarchical clustering as instructed, I need to import the required libraries: `AgglomerativeClustering` from `sklearn.cluster` and `linkage` from `scipy.cluster.hierarchy`. Then, I will prepare to iterate through different linkage criteria, storing the computed linkage matrices.



```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage

linkage_matrices = {}
linkage_criteria = ['ward', 'complete', 'average', 'single']

for criterion in linkage_criteria:
    # AgglomerativeClustering with distance_threshold=0 and n_clusters=None to get full hierarchy
    # The actual fitting is not strictly necessary here for computing the linkage matrix with scipy,
    # but it demonstrates how one would use AgglomerativeClustering.
    # For direct linkage matrix calculation, scipy's linkage function is used on the data.
    agglomerative = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage=criterion)
    agglomerative.fit(df_scaled)

    # Compute the linkage matrix using scipy.cluster.hierarchy.linkage
    # 'method' parameter corresponds to 'linkage' in AgglomerativeClustering
    Z = linkage(df_scaled, method=criterion)
    linkage_matrices[criterion] = Z
    print(f"Linkage matrix for '{criterion}' criterion computed successfully.")
```

**Reasoning**:
Now that the linkage matrices have been computed for each criterion, the next step is to visualize these hierarchical clustering results using dendrograms. This will help us understand the clustering structure for each linkage method.



```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(20, 15))

for i, (criterion, Z) in enumerate(linkage_matrices.items()):
    plt.subplot(2, 2, i + 1)
    dendrogram(Z, truncate_mode='lastp', p=30, show_leaf_counts=True)
    plt.title(f'Dendrogram for {criterion.capitalize()} Linkage')
    plt.xlabel('Cluster size or observation (idx)')
    plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

### Parameter Estimation for DBSCAN

DBSCAN requires two main parameters:

1.  **`eps` (epsilon)**: The maximum distance between two samples for one to be considered as in the neighborhood of the other.
2.  **`min_samples`**: The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself.

#### Estimating `min_samples`
A common heuristic for `min_samples` is to use `2 * number_of_features`.


```python
num_features = df_scaled.shape[1]
min_samples = 2 * num_features
print(f"Estimated min_samples for DBSCAN: {min_samples}")
```

#### Estimating `eps` using K-distance Plot

To find an appropriate `eps` value, we can use the k-distance graph. The idea is to calculate the distance to the `min_samples`-th nearest neighbor for each point and then plot these distances in ascending order. The 'elbow' in this plot often indicates a good `eps` value.


```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Calculate the distance to the k-th nearest neighbor
# k = min_samples

neigh = NearestNeighbors(n_neighbors=min_samples)
neigh.fit(df_scaled)
distances, indices = neigh.kneighbors(df_scaled)

# Sort the distances and plot them
distances = np.sort(distances[:, min_samples-1], axis=0)

plt.figure(figsize=(12, 6))
plt.plot(distances)
plt.xlabel('Points sorted by distance')
plt.ylabel(f'{min_samples}-th Nearest Neighbor Distance (epsilon)')
plt.title('K-distance Graph for DBSCAN Epsilon Estimation')
plt.grid(True)
plt.show()
```

## Apply DBSCAN

Apply the DBSCAN clustering algorithm to the scaled dataset using an estimated `eps` value from the k-distance plot and the calculated `min_samples`.


**Reasoning**:
Now that the `min_samples` and an appropriate `eps` value have been estimated from the k-distance plot, I will apply the DBSCAN clustering algorithm to the scaled dataset.



```python
from sklearn.cluster import DBSCAN

# Instantiate DBSCAN with the estimated parameters
dbsc = DBSCAN(eps=0.6, min_samples=min_samples)

# Fit DBSCAN to the scaled data and get cluster labels
clusters = dbsc.fit_predict(df_scaled)

# Add the cluster labels to the original DataFrame
df['DBSCAN_cluster'] = clusters

print("DBSCAN clustering applied successfully with eps=0.6 and min_samples=24.")
print("Number of clusters found (excluding noise, label -1):", len(set(clusters)) - (1 if -1 in clusters else 0))
print("Number of noise points (label -1):", list(clusters).count(-1))
```

## Evaluate DBSCAN Clustering

### Subtask:
Evaluate the results of the DBSCAN clustering, considering metrics like the number of clusters formed and the number of noise points. Experiment with different `eps` values around the 'elbow' point of the k-distance graph.


```python
print("Distribution of DBSCAN clusters:")
print(pd.Series(clusters).value_counts())

print("\nExperimenting with different eps values:")

# Experiment with different eps values around the 'elbow' point (around 0.6)
eps_values = [0.5, 0.7, 0.8]

for eps in eps_values:
    dbsc_experiment = DBSCAN(eps=eps, min_samples=min_samples)
    clusters_experiment = dbsc_experiment.fit_predict(df_scaled)

    n_clusters_experiment = len(set(clusters_experiment)) - (1 if -1 in clusters_experiment else 0)
    n_noise_experiment = list(clusters_experiment).count(-1)

    print(f"\nResults for eps={eps}:")
    print(f"  Number of clusters: {n_clusters_experiment}")
    print(f"  Number of noise points: {n_noise_experiment}")
    print("  Cluster distribution:")
    print(pd.Series(clusters_experiment).value_counts())
```
